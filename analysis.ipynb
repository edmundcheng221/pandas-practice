Bar Chart of the airline and the corresponding mean flight distance.

```python
df_to_plot = df[ ['AIRLINE','DISTANCE'] ].groupby(by=['AIRLINE']).aggregate({'mean'})
df_to_plot.plot.bar(ylabel='Average Distance', legend=False)
```

Line Plot of the airtime total each month.

```python
ts = df.groupby('MONTH')['AIR_TIME'].count()
ts.plot()
```

Bar Chart showing the Airlines and the number of flights each month.

```python
df_to_plot2 = df[ ['AIRLINE','MONTH'] ].groupby(by=['AIRLINE']).count()
df_to_plot2.plot.bar(ylabel='MONTH', legend=False)
```

Line Chart showing the culmulative sum of airlines flown each month.

```python
ts = ts.cumsum()
ts.plot()
```

Pie Chart showing percentage of distance traveled each month by all airlines.

```python
PiePlot = df.groupby('MONTH')['DISTANCE'].sum()
plotResult = PiePlot.plot.pie(y='DISTANCE', figsize=(5, 5))
```
